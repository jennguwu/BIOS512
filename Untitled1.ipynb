```r
library('tidyverse')
library('ggrepel')
```

```r
wheel = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-08-09/wheels.csv')
```

```r
head(wheel)
```

```r
p = ggplot(wheel, aes(y = country))

p = p + geom_bar(fill = 'steelblue') + labs(title = 'Ferris Wheels by Country', y = 'Country') 

p 
```

In this graphical presentation I have explored ferris wheels by country by using a bar chart. In this bar chart we can see that the U.S. has the most ferris wheels, followed by Japan and China. If you are a fan of ferris wheels then the U.S. is the right place to find one!

```r
options(repr.plot.width = 6, repr.plot.height = 4)

p = ggplot(wheel, aes(x = height, y = diameter, color = type))

p = p + geom_point() + labs(title = 'Ferris Wheel Size by Type', x = 'Diameter', y = 'Height')

p = p + scale_color_brewer(type = "qual", palette = "Set2")

p
```

In this graphical presentation I have explored ferris wheel size by type. We can see here that the largest ferris wheels are of the fixed type. As the name suggests, fixed type are designed for permanent instalation. 

```r
options(repr.plot.width = 8, repr.plot.height = 6)

p = ggplot(wheel, aes(x = height, 
                     y = diameter))

p = p + geom_point(position = position_jitter(h = 0.5, w = 0.25), size = 1) + labs(title = 'Ferris Wheel Size by Status', x = 'Diameter', y = 'Height')

p = p + facet_wrap(~status, nrow = 2)

p
```

In this graphical presentation I have used facet_wrap to examine status of ferris wheels by size. We can see that most ferris wheels are operational, however the 2 largest ferris wheels are not currently available as one is in development and one is delayed.
